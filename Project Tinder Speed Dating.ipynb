# Who are the people in the speed dating experiment ?

```python
# New names for the columns

speed_dating["gender"] = speed_dating['gender'].apply(lambda x: "Female" if x == 0
                                            else 'Male') 

speed_dating["match"] = speed_dating['match'].apply(lambda x: "Match" if x == 1
                                            else "No match") 

speed_dating["race"] = speed_dating['race'].apply(lambda x: "Black/African American" if x == 1
                                        else "European/Caucasian-American" if x == 2 
                                        else "Latino/Hispanic American" if x == 3
                                        else "Asian/Pacific Islander/Asian-American" if x == 4
                                        else "Native American" if x == 5
                                        else "Other" if x == 6
                                        else "nan")

speed_dating["field_cd"] = speed_dating["field_cd"].apply(lambda x: "Law" if x == 1
                                        else "Math" if x == 2 
                                        else "Social Science, Psychologist" if x == 3
                                        else "Medical Science, Pharmaceuticals, and Bio Tech" if x == 4
                                        else "Engineering" if x == 5
                                        else "English/Creative Writing/ Journalism" if x == 6
                                        else "History/Religion/Philosophy" if x == 7
                                        else "Business/Econ/Finance" if x == 8
                                        else "Education, Academia" if x == 9
                                        else "nan")  
```

### What makes people interested into each other to go on a second date together ?

```python
#New dataset with waves 6 to 9 deleted because rating system is different (10 instead of 100 points)
df_less_waves = speed_dating_per_person[(speed_dating_per_person['wave'] != 6) & (speed_dating_per_person['wave'] != 7) & (speed_dating_per_person['wave'] != 8) & (speed_dating_per_person['wave'] != 9)]
df_less_waves.head()
```

### QUESTION 1 What are the least desirable attributes in a male partner? Does this differ for female partners?

```python
attributes_potential_dates = ['gender','attr1_1', 'sinc1_1','intel1_1','fun1_1','amb1_1','shar1_1'] 
attributes_df = df_less_waves[attributes_potential_dates].groupby('gender').mean().dropna() 
#.reset_index(drop=True)  
attributes_df.head(20) 
```

```python
print(attributes_df.isnull().sum()) #ok no nan for attributes
```

```python

attributes_names = ['Attractiveness','Sincerity','Intelligence','Fun','Ambition','Shared Interests']
fig = go.Figure(
    data=[go.Bar(name='Men', x=attributes_names, y=list(attributes_df.iloc[1])),
    go.Bar(name='Women', x=attributes_names, y= list(attributes_df.iloc[0]) )]
)
fig.update_layout(
    barmode='group', 
    title=go.layout.Title(text="What are you looking for in the opposite sex?"),
    title_x = 0.5
)
fig.show()
```

What are the least desirable attributes in a male partner? Does this differ for female partners?
- The least desirable attribute in male and female partners is Ambition. It is the same for male and female partners.
- The most desirable attribute for a male's partner is Attractiveness (around 29 points). Then the second group much lower is, respectively, compounded of Intelligence (around 19 points), Fun(around 17 points) and Sincerity (around 16 points). The second to last is shared interests with 10 points.
- The most desirable attribute for a male's partner is Intelligence (around 21 points). Then the second group is, respectively, compounded of Attractiveness (around 19 points), Sincerity (around 18 points) and Fun (around 17 points). The second to last is shared interests with 12 points.


### QUESTION 2 How important do people think attractiveness is in potential mate selection vs. its real impact?

```python
# % of missing values per column 
(speed_dating_per_person.isnull().sum()/len(speed_dating)).sort_values(ascending=False)
```

```python
attributes_comparison = ['gender','attr4_1','sinc4_1','intel4_1','fun4_1','amb4_1','shar4_1','attr7_2','sinc7_2','intel7_2','fun7_2','amb7_2','shar7_2'] 
attributes_comp_df = df_less_waves[attributes_comparison].groupby('gender').mean().dropna() 
attributes_comp_df.head(20)
```

```python
attributes_names = ['Attractiveness','Sincerity','Intelligence','Fun','Ambition','Shared Interests']
male_attributes = attributes_comp_df.loc["Male"]
male_attributes_thoughts = male_attributes[['attr4_1', 'sinc4_1','intel4_1','fun4_1','amb4_1','shar4_1']]
male_attributes_real = male_attributes[['attr7_2', 'sinc7_2','intel7_2','fun7_2','amb7_2','shar7_2']]

fig = go.Figure(
    data=[go.Bar(name='Thoughts', x=attributes_names, y=list(male_attributes_thoughts), marker_color="lightcoral"),
    go.Bar(name='Real impact', x=attributes_names, y= list(male_attributes_real), marker_color="cornflowerblue" )]
)
fig.update_layout(
    barmode='group', 
    title=go.layout.Title(text="How important do MEN think attractiveness is in potential mate selection vs. its real impact?"),
    title_x = 0.5
)
fig.show()
```

How important do MEN think attractiveness is in potential mate selection vs. its real impact?
- In fact, attractiveness is a little more important in men's partner selection than they thought (almost 36 points in their actual decision vs 34 points in what they they think fellow men look for in women). 
- In general, what men think their fellow men look for in women is quite close of the real impact on their partner  selection decision (maximum only 2 points of difference).

```python
attributes_names = ['Attractiveness','Sincerity','Intelligence','Fun','Ambition','Shared Interests']
female_attributes = attributes_comp_df.loc["Female"]
female_attributes_thoughts = female_attributes[['attr4_1', 'sinc4_1','intel4_1','fun4_1','amb4_1','shar4_1']]
female_attributes_real = female_attributes[['attr7_2', 'sinc7_2','intel7_2','fun7_2','amb7_2','shar7_2']]

fig = go.Figure(
    data=[go.Bar(name='Thoughts', x=attributes_names, y=list(female_attributes_thoughts), marker_color="lightcoral"),
    go.Bar(name='Real impact', x=attributes_names, y= list(female_attributes_real), marker_color="cornflowerblue" )]
)
fig.update_layout(
    barmode='group', 
    title=go.layout.Title(text="How important do WOMEN think attractiveness is in potential mate selection vs. its real impact?"),
    title_x = 0.5
)
fig.show()
```

How important do WOMEN think attractiveness is in potential mate selection vs. its real impact?
- In fact, attractiveness is a little less important in women's partner selection than they thought (almost 28 points in their actual decision vs 30 points in what they they think fellow women look for in men). 
- In general, what women think their fellow men look for in men is quite close of the real impact on their partner  selection decision (maximum only 2 points of difference except for ambition: 5 points of difference).

### QUESTION 3 Are shared interests more important than a shared racial background?

```python
#"imprace": How important is it to you (on a scale of 1-10) that a person you date be of the same racial/ethnic background?
fig = px.histogram (speed_dating_per_person, x="imprace")
fig.show() 
```

```python
sns.countplot(x="imprace", data=speed_dating_per_person);
```

Sharing same race with a potential partner is not that important to people (59% of people gave this criteria 3 or less on a total of 10 points).

```python
sns.kdeplot(data= speed_dating_per_person, x= "int_corr", hue="match")
# People who match have a interest correlation coeficient slightly higher than people who do not match.
```

```python
speed_dating_per_person.groupby('match', as_index=False)[["int_corr","imprace"]].mean()
```

Are shared interests more important than a shared racial background?
- People who match have a interest correlation coeficient slightly higher than people who do not match.
- People who match are less interested in sharing the same race than people who do not match.
- Yes shared interests are more important than a shared racial background.
